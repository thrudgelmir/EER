```python
from matplotlib import pyplot as plt
import matplotlib.gridspec as gridspec
import numpy as np
inner_rounds = 200
def show_plot(d1,d2,d3,d4,orig1,orig2) :
    fig = plt.figure(figsize=(18, 3))
    gs = gridspec.GridSpec(1, 2, width_ratios=[1, 1], wspace=0.1)
    ax_left = plt.subplot(gs[0])
    gs_left = gridspec.GridSpecFromSubplotSpec(1, 2, subplot_spec=gs[0], wspace=0.2)
    ax_left.set_title("hide & seek",pad=20)
    ax_left.axis("off")

    ax_right = plt.subplot(gs[1])
    gs_right = gridspec.GridSpecFromSubplotSpec(1, 2, subplot_spec=gs[1], wspace=0.2)
    ax_right.set_title("wesad",pad=20)
    ax_right.axis("off")

    ax1 = plt.subplot(gs_left[0])
    ax2 = plt.subplot(gs_left[1])
    ax3 = plt.subplot(gs_right[0])
    ax4 = plt.subplot(gs_right[1])

    x = np.arange(0,inner_rounds+1,5)
    y1 = (d1-orig1)*100
    y2 = (d2-orig1)*100
    y3 = (d3-orig2)*100
    y4 = (d4-orig2)*100

    ax1.plot(x,y1,marker='o',markersize=3,color='b',linewidth=1.5)
    ax1.set_xlabel('rounds')
    ax1.set_ylabel('acc improved(%)')
    ax1.set_title('10 epoch')

    ax2.plot(x,y2,marker='o',markersize=3,color='r',linewidth=1.5)
    ax2.set_xlabel('rounds')
    ax2.set_ylabel('acc improved(%)')
    ax2.set_title('15 epoch')

    ax3.plot(x,y3,marker='o',markersize=3,color='b',linewidth=1.5)
    ax3.set_xlabel('rounds')
    ax3.set_ylabel('acc improved(%)')
    ax3.set_title('10 epoch')

    ax4.plot(x,y4,marker='o',markersize=3,color='r',linewidth=1.5)
    ax4.set_xlabel('rounds')
    ax4.set_ylabel('acc improved(%)')
    ax4.set_title('15 epoch')
    
    plt.tight_layout()
    plt.show()
    print(f'max acc improve given\t 10 epochs : {np.max(y1):.4f}%  15 epochs : {np.max(y2):.4f}%')
    print(f'max acc improve wesad\t 10 epochs : {np.max(y3):.4f}%  15 epochs : {np.max(y4):.4f}%')
    print(f'max acc pos given\t 10 epochs : {x[np.argmax(y1)]}  15 epochs: {x[np.argmax(y2)]}')
    print(f'max acc pos wesad\t 10 epochs : {x[np.argmax(y3)]}  15 epochs: {x[np.argmax(y4)]}')
```

```python
import joblib

orig1 = joblib.load('./tmpdir/given_mean_acc.pkl')
orig2 = joblib.load('./tmpdir/wesad_mean_acc.pkl')

d1 = np.mean(joblib.load('./tmpdir/given10.pkl'),axis=0)
d2 = np.mean(joblib.load('./tmpdir/given15.pkl'),axis=0)
d3 = np.mean(joblib.load('./tmpdir/wesad10.pkl'),axis=0)
d4 = np.mean(joblib.load('./tmpdir/wesad15.pkl'),axis=0)

show_plot(d1,d2,d3,d4,orig1,orig2)
```
